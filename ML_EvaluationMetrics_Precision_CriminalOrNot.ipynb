```python
crime = pd.read_csv('https://raw.githubusercontent.com/insaid2018/Term-2/master/CaseStudy/criminal_train.csv')
crime.head()
```

```python
x = crime.drop('Criminal',axis = 1)
y = crime.Criminal

#=== TTS ===
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

#== (1) Model with RF ===
model_rf = RandomForestClassifier(random_state = 0)
model_rf.fit(X_train, y_train)

#predicting on test data
# Storing the results of prediction into another data frame, y_pred_test
y_pred_test_rf = model_rf.predict(X_test)

# Eval metrics
print('Accuracy score for test data is:', accuracy_score(y_test,y_pred_test_rf))
print('Precision score for test data is:', precision_score(y_test,y_pred_test_rf))
print('Recall score for test data is:', recall_score(y_test,y_pred_test_rf))
print('F1 score for test data is:', f1_score(y_test,y_pred_test_rf))

# Confusion Matrix
confusion_matrix_rf = pd.DataFrame(confusion_matrix(y_test, y_pred_test_rf))
confusion_matrix_rf.index = ['Actual Innocent','Actual Criminal']
confusion_matrix_rf.columns = ['Predicted Innocent','Predicted Criminal']
print(confusion_matrix_rf)

```

```python
# === (2) Model with LR ===
# == Just for Fun! ===

from sklearn.linear_model import LogisticRegression 

model_lr = LogisticRegression()
model_lr.fit(X_train,y_train)
y_pred_test_lr = model_lr.predict(X_test)

# Eval metrics
print('Accuracy score for test data is:', accuracy_score(y_test,y_pred_test_lr))
print('Precision score for test data is:', precision_score(y_test,y_pred_test_lr))
print('Recall score for test data is:', recall_score(y_test,y_pred_test_lr))
print('F1 score for test data is:', f1_score(y_test,y_pred_test_lr))

# Confusion Matrix
confusion_matrix_lr = pd.DataFrame(confusion_matrix(y_test, y_pred_test_lr))
confusion_matrix_lr.index = ['Actual Innocent','Actual Criminal']
confusion_matrix_lr.columns = ['Predicted Innocent','Predicted Criminal']
print(confusion_matrix_lr)


```

```python
## ==== MODEL COMPARSIONS ===

x = crime.drop('Criminal',axis = 1)
y = crime.Criminal

#=== TTS ===
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

###### == (1) Model with RF === #######
model_rf_1 = RandomForestClassifier(random_state = 0)
model_rf_1.fit(X_train, y_train)

#predicting on test data
# Storing the results of prediction into another data frame, y_pred_test
y_pred_test_rf_1 = model_rf_1.predict(X_test)

print('########## MODEL 1, Baseline RF ##########')
# Eval metrics
print('Accuracy score for test data is:', accuracy_score(y_test,y_pred_test_rf_1))
print('Precision score for test data is:', precision_score(y_test,y_pred_test_rf_1))
print('Recall score for test data is:', recall_score(y_test,y_pred_test_rf_1))
print('F1 score for test data is:', f1_score(y_test,y_pred_test_rf_1))

print('============================================\n')

# Confusion Matrix
confusion_matrix_rf_1 = pd.DataFrame(confusion_matrix(y_test, y_pred_test_rf_1))
confusion_matrix_rf_1.index = ['Actual Innocent (y=0)','Actual Guilty (y=1)']
confusion_matrix_rf_1.columns = ['Predicted Innocent (y=0)','Predicted Guilty (y=1)']
print(confusion_matrix_rf_1)

###### === (2) Model with RF with hyperparameter tuning ===

model_rf_2 = RandomForestClassifier(
                                    random_state = 0, 
                                    max_depth=5,
                                    min_samples_leaf=5,
                                    min_samples_split=7,
                                    min_weight_fraction_leaf=0.0,
                                    n_estimators=12,
                                    n_jobs=-1
                                  )
                                    
model_rf_2.fit(X_train, y_train)

#predicting on test data
# Storing the results of prediction into another data frame, y_pred_test
y_pred_test_rf_2 = model_rf_2.predict(X_test)

print('\n\n########## MODEL 2, RF with Hyperparameter tuning ##########')
# Eval metrics
print('Accuracy score for test data is:', accuracy_score(y_test,y_pred_test_rf_2))
print('Precision score for test data is:', precision_score(y_test,y_pred_test_rf_2))
print('Recall score for test data is:', recall_score(y_test,y_pred_test_rf_2))
print('F1 score for test data is:', f1_score(y_test,y_pred_test_rf_2))

print('============================================\n')

# Confusion Matrix
confusion_matrix_rf_2 = pd.DataFrame(confusion_matrix(y_test, y_pred_test_rf_2))
confusion_matrix_rf_2.index = ['Actual Innocent (y=0)','Actual Guilty (y=1)']
confusion_matrix_rf_2.columns = ['Predicted Innocent (y=0)','Predicted Guilty (y=1)']
print(confusion_matrix_rf_2)


print('\n\n=======CONCLUSION=========\n')
print('Model 2 will be used because the FP (actually innocent, predicted guilty) is 0 (and is < FP of Model 1, which is 208)')
print('Therefore, the Precision of Model 2 is 100% which is > than Model 1\'s Precision, which is 67%')
print('\n\n=======FIN=========\n\n')
```
